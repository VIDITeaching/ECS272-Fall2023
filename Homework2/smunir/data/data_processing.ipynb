```python
import pandas as pd
```

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='ISO-8859-1')
```

```python
df[['iyear', 'imonth', 'iday', 'country_txt', 'nkill', 'gname', 'location']]
```

# Bar Plot

```python
df_group = df.gname.value_counts()[:20].reset_index().rename(columns={'index': 'group', 'gname': 'count'})
```

```python
df.success.value_counts()
```

```python
df_group['successful'] = df[df.gname.isin(df_group.group)].groupby('gname').success.sum().values
```

```python
df_group['unsuccessful'] = df[df.gname.isin(df_group.group)].groupby('gname').success.count().values - df_group['successful']
```

```python
df_group[['group', 'successful', 'unsuccessful']].to_csv('groups.csv', index=False)
```

# StreamGraph

```python
countries = df.country_txt.value_counts()[:10].reset_index().rename(columns={'index': 'country', 'country_txt': 'count'}).country.tolist()
```

```python
countries
```

```python
df_countries = pd.DataFrame()

df_countries['iyear'] = df.iyear.unique()
```

```python
df_countries
```

```python
for country in countries:
  df_temp = df[df.country_txt == country].groupby('iyear').count().reset_index()[['iyear', 'eventid']].rename(columns={'eventid': country})
  df_countries = df_countries.merge(df_temp, on='iyear', how='left')
  
```

```python
df_countries.fillna(0, inplace=True)
```

```python
df_countries.to_csv('countries.csv', index=False)
```

# Area chart

```python
attack_types = df.attacktype1_txt.unique().tolist()
```

```python
attack_types
```

```python
df_attack_types = pd.DataFrame()

df_attack_types['iyear'] = df.iyear.unique()

df_attack_types
```

```python
for attack_type in attack_types:
  df_temp = df[df.attacktype1_txt == attack_type].groupby('iyear').count().reset_index()[['iyear', 'eventid']].rename(columns={'eventid': attack_type})
  df_attack_types = df_attack_types.merge(df_temp, on='iyear', how='left')

df_attack_types.fillna(0, inplace=True)
```

```python
df_attack_types.to_csv('attack_types.csv', index=False)
```
